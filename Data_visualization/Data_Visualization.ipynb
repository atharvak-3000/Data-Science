#Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title("First")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
#subplots
plt.subplot(2,2,1)
plt.plot(x,x**2)
plt.subplot(2,2,2)
plt.plot(x,y)#will work in 2nd plot
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,x)
```

```python
#Object Oriented Way
fig = plt.figure(figsize=(2,4),dpi=100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.4,.8])
axis2.plot(x,y)
```

```python
#types of plots
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
#boxplot
plt.boxplot(x)
```

```python
#saving plots
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```

#categorical plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
df =sns.load_dataset("tips")
```

```python
df.head()
```

```python
#countplot
sns.countplot(x = df['sex'],hue = df['smoker'])
```

```python
#barplot
import numpy as np

sns.barplot(x=df['sex'],y=df['tip'],estimator=np.sum)
```

```python
#box plot or violinplot
sns.boxplot(x='tip',y = 'day',data=df,palette='rainbow')
#show distribution of categorical data
#gives 5 number summary
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data=df,palette = 'rainbow')
```

```python
#strip plot -- draw scatter plot with one varible is categorical
#swarm plot -- same but points are adjusted to prevent overlap
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
#combining violin and swarm plot
sns.violinplot(x='tip',y='day',data=df)
sns.swarmplot(x='tip',y='day',data=df)
```

```python
#matrix plots
#heat map
flights = sns.load_dataset('flights')
flights.head()
```

```python
import seaborn  as sns
import pandas as pd
import cufflinks as cf
from plotly.offline import iplot#giving problems
import cufflinks as cf
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
import plotly.express as px
fig = px.line(tips, y='total_bill')
fig.show(renderer = "colab")
```

```python
tip_by_day = tips.groupby('day',observed=False)['tip'].mean().reset_index()
fig1 = px.bar(tip_by_day,x='day',y='tip')
fig1.show(renderer = "colab")
```

```python
fig2 = px.scatter(tips, x='total_bill', y='tip', title='Tip vs Total Bill',
                  color='day',  # optional: adds color grouping
                  hover_data=['sex', 'time'])  # optional: adds hover info
fig2.show(renderer = "colab")

```

```python
fig3 = px.box(tips, x='day', y='total_bill', title='Total Bill Distribution by Day')
fig3.show(renderer = 'colab')
```
