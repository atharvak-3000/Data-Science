#Feature Extraction


```python
df = pd.read_csv(r'anime.csv')
```

```python
df.head()
```

```python
#make a new column for episode count
df.loc[2]['Title']
```

```python
def extract_episode(txt):
  check = False
  data = ""
  for i in txt :
    if i == ")":
      check = False
      break
    if check == True:
      data+=i
    if i == '(':
      check = True
  return data
```

```python
df["Episodes"] = df['Title'].apply(extract_episode)
df
```

```python
df["Episodes"] =df["Episodes"].str.replace(" eps","")
df
```

```python
df["Episodes"]=df["Episodes"].astype(int)
```

```python
df.loc[0]['Title']
```

```python
def extract_time(txt):
  check = False
  data = ""
  count = 0;
  for i in txt:
    if count == 19:
      check = False
      break
    if check == True:
      data += i
      count+=1
    if i == ")":
      check = True
  return data
```

```python
df["Time-Stamp"] = df["Title"].apply(extract_time)
df
```

```python
import pandas as pd
from dateutil import parser

def calculate_months(time_range: str) -> int:
    """
    Calculate total months between two dates in format 'Mon YYYY - Mon YYYY'.
    Example: 'Oct 2012 - Mar 2013' → 6 months
    """
    # Split start and end
    start_str, end_str = time_range.split(" - ")

    # Parse dates
    start = parser.parse(start_str)
    end = parser.parse(end_str)

    # Calculate months
    months = (end.year - start.year) * 12 + (end.month - start.month) + 1  # +1 to include both months
    return months


df["Months"] = df["Time-Stamp"].apply(calculate_months)
```
